## Concept 1 - Scatter plot with hue

A scatter plot shows the relationship between two numeric columns.

Each dot is one row. The position shows its values on the x and y axes.

Seaborn's `scatterplot` adds a `hue` parameter that colours dots by a categorical column. This lets you see whether the relationship between two numbers differs by group.

```python

plt.title("City population vs area by region")
plt.show()
```

The chart shows each city as one dot, with city area on the x axis, population on the y axis, and colour showing the region.

## Concept 2 - Bar plot

A bar plot shows the average, or another aggregate, of a numeric column for each category in a categorical column.

Seaborn's `barplot` draws one bar per category and adds confidence interval lines by default.

The height of each bar is the mean of the numeric column for that group.

## Concept 3 - Histogram

A histogram shows the distribution of a single numeric column.

It divides the range of values into bins and counts how many rows fall into each bin.

Seaborn's `histplot` draws cleaner histograms than matplotlib's default and can overlay a density curve with `kde=True`.

```python
plt.show()
```

The chart shows how the city population values are spread across smaller and larger cities.

## Concept 4 - Box plot

A box plot shows the spread of a numeric column for each category.

The box covers the middle 50% of values, called the interquartile range. The line inside the box is the median. The whiskers extend to the typical range. Dots beyond the whiskers are potential outliers.

Seaborn's `boxplot` makes it easy to compare distributions across groups.

## Concept 5 - Heatmap

A heatmap shows a grid of values where colour represents magnitude.

In data analysis, heatmaps are most commonly used to show a correlation matrix: how strongly each numeric column relates to every other numeric column.

Values close to 1 or -1 indicate strong relationships. Values close to 0 indicate weak or no relationship. Use `df.corr()` to compute the correlation matrix, then pass it to `sns.heatmap()`.

<div style="border:2px solid #1d4ed8; border-left:8px solid #1d4ed8; padding:12px 14px; border-radius:8px; margin:10px 0;">

Read this example, then type your own answer in the live answer cell below.
</div>

```python
dummy_corr = dummy_df[["population", "area_km2"]].corr()
sns.heatmap(dummy_corr, annot=True, fmt=".2f", cmap="coolwarm")
plt.title("Correlation matrix - city data")
plt.show()
```


The chart shows the correlation between city population and area, with colour and numbers both showing the strength of the relationship.
