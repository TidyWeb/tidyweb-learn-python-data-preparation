## 1. Check the Python tools

A **library** is a packaged collection of tools. `import` makes a library available in this notebook. The code below prints the versions that are running.


```python
import sys
import pandas as pd
import matplotlib
import seaborn as sns
import scipy

print('Python:', sys.version.split()[0])
print('pandas:', pd.__version__)
print('matplotlib:', matplotlib.__version__)
print('seaborn:', sns.__version__)
print('scipy:', scipy.__version__)
```

## 2. Check a supplied CSV

A **CSV** is a plain-text data file with values separated by commas. `Path` helps us refer to a file relative to this notebook, so no personal computer path is needed.


```python
from pathlib import Path

dataset_path = Path('../datasets/iris.csv')
print('Dataset found:', dataset_path.exists())
print('Dataset path:', dataset_path)

if not dataset_path.exists():
    raise FileNotFoundError('Keep this notebook inside notebooks/ and keep datasets/ beside it in the project folder.')
```

## 3. Read the first rows

`pd.read_csv(...)` asks the pandas toolbox to turn the CSV into a DataFrame: a table with rows and columns.


```python
df = pd.read_csv(dataset_path)
print('Rows and columns:', df.shape)
df.head()
```
