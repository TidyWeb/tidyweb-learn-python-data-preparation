## Concept 1 - What Statistics Are For

Statistics are shortcut descriptions of a table. They help you answer questions like: what is a typical value here, how spread out are the values, and what shape does the column seem to have? A statistic is useful because it compresses many rows into a smaller summary, but that same compression means it cannot tell the whole story on its own.


## Concept 2 - Mean

The mean is the average. It gives you one central value by using every number in the column. This is useful when you want a quick sense of the general level of the numbers.

Worked example breakdown:

- Full command: `dummy_df["amount"].mean()`
- `dummy_df` is the neutral dummy table.
- `[` and `]` are the brackets used to reach into one named column.
- `"amount"` is the column label written inside quotes.
- `.mean()` is the pandas tool that calculates the average of the selected column.


```python
dummy_df["amount"].mean()
```

## Concept 3 - Median

The median is the middle value after the numbers have been put in order. It is useful when you want a centre measure that is less affected by very large or very small values.

Worked example breakdown:

- Full command: `dummy_df["amount"].median()`
- `dummy_df` names the dummy table.
- `"amount"` names the numeric column we want to inspect.
- `.median()` asks for the middle value after sorting that column.


```python
dummy_df["amount"].median()
```

## Concept 4 - Mode

The mode means the most common value. This can return more than one value if several values are tied for being most common. That matters because mode is about frequency, not about the middle or the average.

Worked example breakdown:

- Full command: `dummy_df["category"].mode()`
- `dummy_df` is the dummy table.
- `"category"` names the column whose repeated values we want to inspect.
- `.mode()` asks pandas for the value or values that appear most often.
- The result can contain more than one row because more than one value can share the top frequency.


```python
dummy_df["category"].mode()
```

## Concept 5 - Standard Deviation

Spread means how tightly packed together or how widely scattered the values are. Standard deviation is one way to describe that spread. A smaller standard deviation usually means the values sit closer together. A larger standard deviation usually means they are more spread out.

Worked example breakdown:

- Full command: `dummy_df["score"].std()`
- `dummy_df` is the dummy table.
- `"score"` names the numeric column whose spread we want to describe.
- `.std()` asks pandas for the standard deviation of that column.
- The answer is one number describing spread, not a picture of the whole distribution.


```python
dummy_df["score"].std()
```

## Concept 6 - Distribution Shape

A distribution shape is the overall pattern of how values are spread across a column. A histogram helps you see whether values bunch together, stretch out, or lean more to one side.

Worked example breakdown:

- Full command: `dummy_df["score"].plot(kind="hist", bins=5)`
- `dummy_df` is the dummy table.
- `"score"` names the numeric column to draw.
- `.plot(...)` asks pandas to create a chart from that column.
- `kind="hist"` tells pandas that the chart type should be a histogram.
- `bins=5` splits the number line into five bar ranges.


```python
dummy_df["score"].plot(kind="hist", bins=5)
plt.title("Score distribution - dummy statistics data")
plt.show()
```

## Concept 7 - Sample Vs Population

A population means every case you care about. A sample means only some of those cases. Most real datasets should be treated as samples unless you know they contain every relevant case. That matters because a summary statistic always describes the table you have directly, but it does not automatically describe a wider world with certainty.
