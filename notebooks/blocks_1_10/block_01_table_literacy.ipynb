This concept only reads and displays the table; the next concept teaches storing that table under the name `df`.

### Command Panel

## Concept 6 - Summary With df.describe(include="all")

`df.describe(include="all")` calculates a built-in summary for the columns. Numeric columns get rows such as mean, std, min, 25%, 50%, 75%, and max. Text columns get rows such as count, unique, top, and freq. `NaN` usually means that summary slot does not apply to that column type.

### Command Panel

```python
df.describe(include="all")
```

**Plain-English meaning**  
Build a summary table for both numeric and text-style columns.

**Code breakdown**

- `df` = the pandas table.
- `.` = reach into the table.
- `describe` = the summary-building tool.
- `(` = start the tool's options.
- `include="all"` = tell pandas to include both numeric and text-style columns.
- `)` = end the tool's options.

**Worked example - dummy bread stock data**
