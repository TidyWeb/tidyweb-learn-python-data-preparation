## Concept 1 - What A Chart Is And Why It Exists

A chart helps the eye notice pattern, spread, imbalance, or relationship faster than a long printed table. A chart is not automatically an explanation. It is a view of the data from one angle.

## Concept 6 - Chart Choice Rule

Different chart types answer different kinds of questions. A bar chart fits category comparison. A histogram fits one numeric distribution. A scatter plot fits the relationship between two numeric columns.


**Worked dummy example - bar chart chosen for a category question**

```python

plt.show(block=False)
print("A bar chart fits this question because the data is categorical and we want to compare counts.")
```

## Concept 7 - What A Chart Shows And What It Cannot Show

A chart should be followed by two sentences: one saying what the chart shows, and one saying what it cannot show from that view alone. The chart is evidence, not the full explanation.


**Worked dummy example - one show sentence and one limit sentence**

```python
dummy_df["category"].value_counts().plot(kind="bar")
plt.show(block=False)
print("Shows: Category A appears most often in this dummy table.")
print("Cannot show: This chart does not explain why Category A appears most often.")
```
