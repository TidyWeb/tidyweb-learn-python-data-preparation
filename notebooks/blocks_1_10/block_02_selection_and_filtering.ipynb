## Concept 2 - Outer And Inner Bracket Rule

A filtering line contains two jobs nested inside each other. The inner part asks the row-by-row question. The outer part keeps only the rows where the answer is yes.

### Command Panel

```python
dummy_df[dummy_df["category"] == "bread"]
```

**Plain-English meaning**  
From the dummy table, keep only the rows where the `category` column equals `"bread"`.

**Code breakdown**

- `outer dummy_df[ ... ]` = keep rows from the table.
- `inner dummy_df["category"] == "bread"` = ask the row-by-row question.
- `dummy_df["category"]` = look inside one column before comparing its values.
- `==` = equals.
- `"bread"` = the text value being tested.

## Concept 7 - Two-Condition AND Filter With &

An AND filter keeps a row only when both conditions are true. If either condition fails, that row is not kept.

### Command Panel

```python
dummy_df[(dummy_df["category"] == "bread") & (dummy_df["price"] > 1.00)]
```

**Plain-English meaning**  
Keep rows where the `category` is `"bread"` and the `price` is greater than `1.00`.

**Code breakdown**

- `outer dummy_df[ ... ]` = keep rows from the table.
- `(dummy_df["category"] == "bread")` = first condition.
- `&` = both conditions must be true.
- `(dummy_df["price"] > 1.00)` = second condition.
- `( ) around each condition` = group each test clearly before joining them.

<div style="border:2px solid #1d4ed8; background:#eff6ff; padding:12px 14px; border-radius:8px; margin:10px 0;">
<strong>Worked example code cell is directly below.</strong><br>
This blue-marked step is a demonstration. Read it and inspect the output before you answer the Iris question.
</div>


````python
dummy_df[(dummy_df["category"] == "bread") & (dummy_df["price"] > 1.00)]
````

### Command Panel

```python

```

**Plain-English meaning**  
Keep rows where the `category` is `"bread"` or the `price` is greater than `1.00`.

**Code breakdown**

- `outer dummy_df[ ... ]` = keep rows from the table.
- `(dummy_df["category"] == "bread")` = first condition.
- `|` = either condition being true is enough.
- `(dummy_df["price"] > 1.00)` = second condition.
- `( ) around each condition` = group each test clearly before joining them.

## Concept 9 - Parentheses Rule For Combined Conditions

When you combine conditions with `&` or `|`, each condition needs its own parentheses. The parentheses make it clear which test belongs together before the join happens.

### Command Panel

```python
dummy_df[(dummy_df["category"] == "bread") & (dummy_df["count"] > 10)]
```

**Plain-English meaning**  
Keep rows where the `category` is `"bread"` and the `count` is greater than `10`, with each condition wrapped in parentheses.

**Code breakdown**

- `(dummy_df["category"] == "bread")` = one complete condition, wrapped as one group.
- `&` = join the grouped conditions with AND.
- `(dummy_df["count"] > 10)` = the second complete condition, wrapped as one group.
- `outer dummy_df[ ... ]` = keep rows where the grouped test is true.
- `parentheses` = show Python exactly how the combined condition is grouped.

<div style="border:2px solid #1d4ed8; background:#eff6ff; padding:12px 14px; border-radius:8px; margin:10px 0;">
<strong>Worked example code cell is directly below.</strong><br>
This blue-marked step is a demonstration. Read it and inspect the output before you answer the Iris question.
</div>


````python
dummy_df[(dummy_df["category"] == "bread") & (dummy_df["count"] > 10)]
````

## Concept 10 - Check Filtered Result Size With .shape

After filtering, you can ask how large the filtered table is. `shape` reports rows first and columns second, and it does not use `()` because it is a stored property, not a called tool.

### Command Panel

```python
dummy_df[dummy_df["category"] == "bread"].shape
```

**Plain-English meaning**  
Filter the table first, then show the size of the filtered result as `(rows, columns)`.

**Code breakdown**

- `dummy_df[dummy_df["category"] == "bread"]` = make the smaller filtered table first.
- `.` = reach into that filtered result.
- `shape` = the stored size property.
- `no ()` = no brackets because `shape` is a property, not a called tool.
- `(rows, columns)` = the order of the size result.

<div style="border:2px solid #1d4ed8; background:#eff6ff; padding:12px 14px; border-radius:8px; margin:10px 0;">
