## 1. A list keeps items in order

A list is one value that holds several items in a chosen order. Square brackets make the list. Positions start at zero, so the first item has position `0`.


### Command bridge

```python
routes = ["River", "Hill", "Town"]
```

**Plain-English meaning:** `routes` names the collection; `[` and `]` hold its items; commas separate items; the quoted route names are strings.


### Worked example — dummy material

## 2. A dictionary is one labelled record

A dictionary stores values under meaningful keys instead of positions. Curly brackets make the record. A key is a label; a value is the information under that label.


### Command bridge

```python
volunteer = {"name": "Ava", "bags": 3}
```

**Plain-English meaning:** `volunteer` is the record name; `{}` make a dictionary; `"name"` and `"bags"` are keys; the values follow each colon.


### Worked example — dummy material

This is not your lesson material. It is only a safe place to watch the move work.


````python
book = {"title": "North Wind", "copies": 4}
print(book["title"])
````

## 3. Change one field by key

A record can be updated by writing a new value under an existing key. This is useful when a fact changes while the rest of the record stays the same.


### Command bridge

```python
volunteer["bags"] = 5
```

**Plain-English meaning:** `volunteer["bags"]` points to one field; `=` replaces the value held in that field; `5` is the new value.


### Worked example — dummy material

This is not your lesson material. It is only a safe place to watch the move work.


````python
book = {"title": "North Wind", "copies": 4}
book["copies"] = 5
print(book)
````

## 4. A list of records resembles a small table

A list can hold dictionaries with the same keys. Each dictionary acts like one row; each repeated key acts like a column name. This is the bridge to tables later.


### Command bridge

```python
volunteers = [{"name": "Ava", "bags": 3}, {"name": "Noah", "bags": 1}]
```

**Plain-English meaning:** `volunteers` names the collection; each `{...}` is one labelled record; the repeated keys make the records comparable.


### Worked example — dummy material

This is not your lesson material. It is only a safe place to watch the move work.


````python
books = [{"title": "North Wind", "copies": 4}, {"title": "Bright Lake", "copies": 1}]
print(books[1]["title"])
````

## 5. A for loop visits every record

A `for` loop repeats an action once for each item in a collection. The loop variable is a temporary name for the current record.


### Command bridge

```python
for volunteer in volunteers:
    print(volunteer["name"])
```

**Plain-English meaning:** `for` starts repetition; `volunteer` is the current-record name; `in` says where items come from; `volunteers` is the collection; the indented print runs once per record.


### Worked example — dummy material

This is not your lesson material. It is only a safe place to watch the move work.


````python
books = [{"title": "North Wind", "copies": 4}, {"title": "Bright Lake", "copies": 1}]
for book in books:
    print(book["title"])
````

## 6. A condition inside a loop selects records

A condition inside a loop asks the same question about every record. That makes it possible to notice which records meet a rule.


### Command bridge

```python
if volunteer["bags"] >= 3:
    print(volunteer["name"])
```

**Plain-English meaning:** `volunteer["bags"]` reads the current value; `>= 3` makes a Boolean; `if` controls the indented action for that record.


### Worked example — dummy material

This is not your lesson material. It is only a safe place to watch the move work.


````python
books = [{"title": "North Wind", "copies": 4}, {"title": "Bright Lake", "copies": 1}]
for book in books:
    if book["copies"] >= 3:
        print(book["title"])
````

## 7. Build a new filtered list

A filtered list keeps selected records for later use. Begin with an empty list, then append records that pass the rule. `append` means add one item to the end.


### Command bridge

```python
busy_routes = []
if volunteer["bags"] >= 3:
    busy_routes.append(volunteer)
```

**Plain-English meaning:** `busy_routes = []` creates an empty list; the `if` decides; `.append(...)` adds the current record when it qualifies.


### Worked example — dummy material

This is not your lesson material. It is only a safe place to watch the move work.


````python
popular = []
for book in [{"title": "North Wind", "copies": 4}, {"title": "Bright Lake", "copies": 1}]:
    if book["copies"] >= 3:
        popular.append(book)
print(popular)
````
