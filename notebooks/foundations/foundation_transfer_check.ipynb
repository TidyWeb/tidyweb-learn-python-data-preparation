# Foundation Transfer Check

This is retrieval practice, not another taught lesson. There are no worked answers, starter code, or hidden answer shapes. Use a new neutral story: a **town noticeboard** records local notices with a title, a category, and a number of days remaining.

Give yourself 30–40 minutes. Use only concepts already learned in Foundations 1–3.


**Your turn — town noticeboard.** Name three literal values you will need and store them in clear variables. Include one text value, one number, and one Boolean.


**Your turn — town noticeboard.** Write an ordered `if` / `elif` / `else` decision that classifies a notice by days remaining. Test at least one boundary value.


**Your turn — town noticeboard.** Create a list containing at least three notice dictionaries. Use the same keys in every record.


**Your turn — town noticeboard.** Loop through the records and build a new list containing only notices that meet your chosen condition.


**Your turn — town noticeboard.** Write and call one function that returns a useful result from one notice or from your selected list.


**Your turn — town noticeboard.** Convert your records to a pandas DataFrame. In a short comment, state what one row and one column mean in this story.


**Your turn — reflection.** Write a short plain-English finding from the data and name one idea you would revisit before starting Block 1.
