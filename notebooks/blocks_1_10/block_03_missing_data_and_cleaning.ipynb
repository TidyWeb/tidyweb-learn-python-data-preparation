## Concept 1 - Find Missing Cells With df.isnull()

Check every cell in the table and return `True` where the value is missing and `False` where the value exists.

- `isnull` = the tool that tests each cell for a missing value.
- `()` = brackets used to call the tool.
- `True` means the cell is missing and `False` means the cell has a value.

## Concept 2 - Count Missing Values With df.isnull().sum()

After you find the missing cells, the next question is how many are missing in each column. This line chains two steps together: first mark missing cells, then count them.

### Command Panel

```python
dummy_df.isnull().sum()
```

**Plain-English meaning**  
Count how many missing values appear in each column of the table.

**Code breakdown**

- `dummy_df` = the table variable.
- `.isnull()` = mark missing cells as `True` and present cells as `False`.
- `.` = pass the result of one step into the next step.
- `.sum()` = add the `True` values column by column. `True` counts as 1 and `False` counts as 0.
- `isnull().sum()` = a chain: find missing cells first, then count them.

## Concept 3 - Make A Safe Copy With df.copy()

Before you change a table, make a separate working copy. That way the original table stays available if you need to compare, restart, or undo a cleaning choice.

### Command Panel

```python

```

**Plain-English meaning**  
Create a separate copy of the table and store it under the name `df_clean`.

**Code breakdown**

- `df_clean` = the name for the new working copy.
- `=` = store the result on the right under the name on the left.
- `dummy_df` = the original dummy table.
- `.copy()` = make a separate copy so later changes do not affect the original table.
- `df_clean` and `dummy_df` are separate after this line runs.

## Concept 4 - Remove Rows With Any Missing Value Using dropna()

A broad cleaning step is to remove any row that contains at least one missing value anywhere in that row. This can be useful, but it can also remove more data than you intended.

### Command Panel

```python

```

**Plain-English meaning**  
Remove every row that contains at least one missing value.

**Code breakdown**

- `df_clean` = the working copy being cleaned.
- `.` = reach into the table copy and use one of its tools.
- `dropna` = the tool that removes rows with missing values.
- `()` = brackets used to call the tool.
- `without subset` = check every column in the row, not just one column.

## Concept 5 - Remove Rows Missing One Named Column With dropna(subset=[...])

Sometimes you only care whether one specific column is missing. The `subset=` argument lets you limit the check to the named column or columns instead of the whole row.

### Command Panel

```python
df_clean.dropna(subset=["shelf"])
```

**Plain-English meaning**  
Remove only the rows where the `shelf` column is missing.

**Code breakdown**

- `df_clean` = the working copy being cleaned.
- `.dropna(...)` = the row-removal tool.
- `subset=` = limit which columns are checked for missing values.
- `["shelf"]` = a list naming the column to check.
- `dropna(subset=[...])` = narrower than plain `dropna()` because it does not check every column.

<div style="border:2px solid #1d4ed8; background:#eff6ff; padding:12px 14px; border-radius:8px; margin:10px 0;">

## Concept 6 - Fill Missing Values In One Column With df["column"].fillna(value)

Instead of removing rows, you can replace missing values with something you choose. Filling one named column is often safer than filling the whole table because the replacement can match that column's meaning.

### Command Panel

```python

```

**Plain-English meaning**  
Replace missing values in the `shelf` column with the word `"Unknown"`.

**Code breakdown**

- `df_clean` = the working copy.
- `["shelf"]` = select one column by name.
- `.` = reach into that selected column and use one of its tools.
- `fillna` = the tool that replaces missing values.
- `("Unknown")` = the replacement value used where the column has `NaN`.

<div style="border:2px solid #1d4ed8; background:#eff6ff; padding:12px 14px; border-radius:8px; margin:10px 0;">

## Concept 8 - Safe Cleaning Pattern: Copy First, Then Clean

The safest cleaning habit is a two-line pattern: first copy the original table, then clean the copy. That keeps the original table untouched while you test your cleaning decision.

### Command Panel

```python
df_clean = df.copy()
df_clean = df_clean.dropna(subset=["shelf"])
```

**Plain-English meaning**  
Make a working copy, then clean the copy by removing rows where one named column is missing.

**Code breakdown**

- `df_clean = df.copy()` = create the separate working copy first.
- `df_clean =` = store the cleaned result back into the working-copy name.
- `df_clean.dropna(...)` = clean the working copy, not the original table.
- `subset=["shelf"]` = say exactly which column controls the removal step.
- `df` stays untouched while `df_clean` changes.

<div style="border:2px solid #1d4ed8; background:#eff6ff; padding:12px 14px; border-radius:8px; margin:10px 0;">
<strong>Worked example code cell is directly below.</strong><br>
This blue-marked step is a demonstration. Read it and inspect the output before you answer the lesson question.
</div>


````python
df_clean = dummy_df.copy()
df_clean = df_clean.dropna(subset=["shelf"])
df_clean
````
