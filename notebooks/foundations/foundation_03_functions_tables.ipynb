## 1. A function names repeated work

A function is a named recipe for a job you may need more than once. Defining it stores the recipe; it does not run the recipe yet.


### Command bridge

```python
def welcome(name):
    return "Welcome, " + name
```

**Plain-English meaning:** `def` begins the definition; `welcome` is the function name; `name` is a parameter; `return` sends a result back; indentation belongs to the function.


### Worked example — dummy material

## 2. Definition and call are different

A definition creates the recipe. A call asks Python to use it now. Parentheses after the name mean a call.


### Command bridge

```python
message = welcome("Mina")
```

**Plain-English meaning:** `message` is where the result is stored; `welcome(...)` calls the function; `"Mina"` is the argument supplied for its parameter.


### Worked example — dummy material

This is not your lesson material. It is only a safe place to watch the move work.


````python
def label_book(title):
    return "Shelf label: " + title
message = label_book("Bright Lake")
print(message)
````

## 3. Parameters receive arguments

A parameter is the placeholder written in a definition. An argument is the actual value supplied in a call. The function can use the parameter’s temporary local name.


### Command bridge

```python
def add_minutes(minutes, extra):
    return minutes + extra
```

**Plain-English meaning:** `minutes` and `extra` are parameters; `return` gives back their sum. In a call, the actual numbers are arguments.


### Worked example — dummy material

This is not your lesson material. It is only a safe place to watch the move work.


````python
def add_pages(pages, bonus):
    return pages + bonus
print(add_pages(120, 8))
````

## 4. Return a useful result

A returned value travels back to the place where the function was called. That lets you store, print, compare, or reuse the result elsewhere.


### Command bridge

```python
average = total / count
```

**Plain-English meaning:** `total` and `count` are values already available; `/` divides; `average` stores the new returned-or-calculated result.


### Worked example — dummy material

This is not your lesson material. It is only a safe place to watch the move work.


````python
def average_pages(total, count):
    return total / count
print(average_pages(180, 3))
````

## 5. Accumulate across records

An accumulator is a variable that starts at a simple value and changes repeatedly. It helps build one total while a loop visits many records.


### Command bridge

```python
total = 0
for repair in repairs:
    total = total + repair["minutes"]
```

**Plain-English meaning:** `total = 0` starts the accumulator; the loop visits each record; `repair["minutes"]` reads one field; the new total replaces the old one.


### Worked example — dummy material

This is not your lesson material. It is only a safe place to watch the move work.


````python
books = [{"title": "North Wind", "pages": 120}, {"title": "Bright Lake", "pages": 180}]
total = 0
for book in books:
    total = total + book["pages"]
print(total)
````

## 6. A DataFrame is a table-shaped object

Pandas is a Python library: a packaged toolbox for table-shaped data. `import pandas as pd` gives that toolbox the short name `pd`. A DataFrame is pandas’ table object; it has rows and named columns.


### Command bridge

```python
import pandas as pd
df = pd.DataFrame(repairs)
```

**Plain-English meaning:** `import` loads a library; `pandas` is the library; `as pd` gives it a short nickname; `pd.DataFrame(...)` builds a table; `repairs` supplies the list of records; `df` stores the table.


### Worked example — dummy material

This is not your lesson material. It is only a safe place to watch the move work.


````python
import pandas as pd
books = [{"title": "North Wind", "pages": 120}, {"title": "Bright Lake", "pages": 180}]
df = pd.DataFrame(books)
print(df)
````
