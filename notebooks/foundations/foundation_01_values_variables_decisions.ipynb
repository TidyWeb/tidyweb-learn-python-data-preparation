## 1. Values and their types

A literal value is written directly in the code. Text is a string inside quotes; a whole number is an integer; a decimal is a float; `True` and `False` are Boolean values. Types matter because Python treats text, numbers, and decisions differently.


### Command bridge

```python

```

**Plain-English meaning:** `guest_limit` is a variable name; `=` stores the value on its right; `40` is an integer value.

## 2. Variables keep values by name

A variable is a chosen label attached to a value. Assignment uses one equals sign: it stores, it does not ask whether two things are equal.


### Command bridge

```python

```

**Plain-English meaning:** `available_seats` is the chosen label; `=` stores; `32` is the value now kept under that label.

## 3. Comparisons make a decision value

A comparison asks a question and produces either `True` or `False`. `>=` means greater than or equal to; it is different from the single `=` used for storage.


### Command bridge

```python
enough_seats = remaining >= 10
```

**Plain-English meaning:** `remaining` supplies a number; `>=` compares it; `10` is the threshold; `enough_seats` stores the Boolean answer.

## 4. One if gate

An `if` gate runs an action only when its condition is `True`. The indented line belongs to the gate; indentation is Python’s visible way of showing what the gate controls.


### Command bridge

```python
if can_accept_group:
    print("Offer the group seats")
```

**Plain-English meaning:** `if` starts the gate; `can_accept_group` is the Boolean condition; `:` opens the controlled block; the indented `print` is the action.


### Worked example — dummy material

This is not your lesson material. It is only a safe place to watch the move work.


````python
has_ticket = True
if has_ticket:
    print("Welcome in")
````

## 5. Ordered choices with if, elif, and else

Use `elif` for another condition only when earlier conditions were false. `else` is the remaining case. The order matters: Python stops at the first true branch.


### Command bridge

```python
if rain_chance >= 70:
    print("Move indoors")
elif rain_chance >= 40:
    print("Check again later")
else:
    print("Screen outside")
```

**Plain-English meaning:** `if` checks first; `elif` means otherwise, check this; `else` means otherwise do the final action. Each indented print belongs to its branch.


### Worked example — dummy material

This is not your lesson material. It is only a safe place to watch the move work.


````python
temperature = 13
if temperature < 10:
    print("Bring heaters")
elif temperature < 16:
    print("Offer blankets")
else:
    print("Comfortable outside")
````

## 6. Combining and reversing conditions

`and` needs both conditions to be true. `or` needs at least one. `not` reverses a Boolean answer. These let one decision reflect more than one real-world rule.


### Command bridge

```python
can_screen = (rain_chance < 40) and outdoor
```

**Plain-English meaning:** The parentheses group the first comparison; `and` requires that answer and `outdoor` to be true; `can_screen` stores the result.


### Worked example — dummy material

This is not your lesson material. It is only a safe place to watch the move work.


````python
is_weekend = True
is_closed = False
can_open = is_weekend and not is_closed
print(can_open)
````
