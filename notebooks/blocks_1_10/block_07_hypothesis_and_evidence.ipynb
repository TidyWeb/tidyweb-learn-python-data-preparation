## Concept 1 - What A Hypothesis Is

A hypothesis is a clear, testable statement about what you expect to find in the data. There are always two parts: the null hypothesis and the alternative hypothesis.

The null hypothesis says there is no meaningful difference or effect. The alternative hypothesis says there is a meaningful difference or effect. You state both before looking at test results. Not after.

**Worked dummy example - state both hypotheses clearly**

```python
print("Null hypothesis: There is no difference in exam scores between Group A and Group B.")
print("Alternative hypothesis: Group A scores higher on average than Group B.")
```

**What each part means**

- `print(...)` = show a sentence clearly beneath the code cell.
- `Null hypothesis` = the no-difference starting position.
- `Alternative hypothesis` = the claim that a difference exists.
- `Group A` and `Group B` = the two groups being compared in the dummy data.

## Concept 2 - Framing A Testable Question From Data

Not every question is testable. A testable question names two groups and one numeric measure. That gives you something concrete to compare.

A question like <strong>Do customers in the Consumer segment spend more on average than Corporate customers?</strong> is testable because it names two groups and one numeric measure. A question like <strong>Is the data interesting?</strong> is not testable because it does not define what should be compared.

To frame a testable question you need a numeric column and two groups taken from a categorical column.

**Worked dummy example - split the groups and compare their means**

```python
group_a = dummy_df[dummy_df["group"] == "A"]["score"]
group_b = dummy_df[dummy_df["group"] == "B"]["score"]
print(group_a.mean())
print(group_b.mean())
```

**What each part means**

- `dummy_df[...]` = filter the dummy table down to rows that match a condition.
- `dummy_df["group"] == "A"` = keep only rows from Group A.
- `dummy_df["group"] == "B"` = keep only rows from Group B.
- `['score']` = select the numeric measure to compare.
- `.mean()` = calculate the average for each group.

## Sequencing Rule - Hypotheses Before Tests

Before any `scipy` test, t-statistic, or p-value appears, the hypothesis must already be stated in plain English. That sequence is not optional.

In this notebook, the hypothesis framing comes first. Only after that do we move to statistical evidence.

<div style="border:2px solid #ca8a04; border-left:8px solid #ca8a04; padding:14px 16px; border-radius:4px; margin:10px 0;">
<strong>Translation:</strong> first decide what claim is being tested, then test it. Do not let the test result invent the question after the fact.
</div>

*No p-value belongs on the page before the hypothesis does.*


## Concept 3 - What A p-value Is

A p-value is a number that measures how surprising your result would be if the null hypothesis were true. It is not a verdict. It is evidence.

A low p-value, commonly below `0.05`, means the observed result would be unusual if there were no real difference. A high p-value means the result is more consistent with there being no real difference. The threshold of `0.05` is a convention, not a law. A p-value does not tell you how large or how important the difference is.

**Worked dummy example - run the t-test and inspect the p-value**

```python
result = stats.ttest_ind(group_a, group_b)
print(result.pvalue)
print(f"A p-value of {result.pvalue:.4f} means that if there were truly no difference between the groups, a result this extreme would occur roughly {result.pvalue * 100:.2f}% of the time by chance alone.")
```

**What each part means**

- `stats` = the `scipy.stats` toolbox imported in setup.
- `.ttest_ind(...)` = the independent-samples t-test.
- `group_a, group_b` = the two numeric groups being compared.
- `result.pvalue` = the probability-style evidence number returned by the test.

## Concept 4 - Running A t-test

A t-test compares the means of two groups to decide whether the difference is likely to be real or likely to be chance. The independent-samples t-test, `ttest_ind`, is used when the two groups have no direct pairing with each other.

It produces a t-statistic and a p-value. The t-statistic measures how far apart the group means are relative to the spread of the data. The p-value tells you how likely that distance would be if the null hypothesis were true.

**Worked dummy example - print the full t-test result**

```python
result = stats.ttest_ind(group_a, group_b)
print("t-statistic:", result.statistic)
print("p-value:", result.pvalue)
```

**What each part means**

- `result.statistic` = the distance-style summary from the test.
- `result.pvalue` = the probability-style evidence from the test.
- The two numbers answer different questions, so both should be read.

<div style="border:2px solid #1d4ed8; border-left:8px solid #1d4ed8; padding:12px 14px; border-radius:8px; margin:10px 0;">
<strong>Worked example code cell is directly below.</strong><br>
This blue-marked step is a demonstration. Read it and inspect the output before you answer the Superstore question.
</div>


````python
result = stats.ttest_ind(group_a, group_b)
print("t-statistic:", result.statistic)
print("p-value:", result.pvalue)
````

## Concept 5 - Statistical Significance Vs Practical Importance

Statistical significance means the result is unlikely to be explained by chance alone. Practical importance means the difference is large enough to matter in the real world. These are not the same question.

A very large dataset can produce a statistically significant result for a difference so small it has little practical meaning. A small dataset can produce a non-significant result even when the practical difference is substantial. Always ask both questions: Is this result statistically significant? And does the size of the difference matter?

**Worked dummy example - compare the mean difference to the p-value**

```python
print(group_a.mean())
print(group_b.mean())
print("Mean difference:", round(group_a.mean() - group_b.mean(), 2))
print("Is this difference meaningful in context? That depends on the domain.")
```

**What each part means**

- `group_a.mean()` and `group_b.mean()` = the average score in each group.
- `group_a.mean() - group_b.mean()` = the size of the gap between those averages.
- The p-value addresses statistical evidence. The mean difference helps you judge practical size.
