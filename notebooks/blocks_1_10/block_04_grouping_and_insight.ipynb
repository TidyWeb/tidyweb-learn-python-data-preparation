## Concept 1 - What Grouping Means

Grouping means splitting the rows into category buckets, then calculating something inside each bucket. Instead of describing one row at a time, you compare groups.

**Worked dummy example - grouped average**

```python
dummy_df.groupby("category")["value"].mean()
```

**What each part means**

- `dummy_df` = the dummy table.
- `.groupby("category")` = split the rows into category groups.
- `["value"]` = choose the column to summarise inside each group.
- `.mean()` = calculate the average inside each group.

## Concept 4 - Group And Average With df.groupby(...)[...].mean()

A grouped average compares the typical value inside each group. The group column creates the buckets. The value column tells pandas what to average inside each bucket.

**Worked dummy example - average by group**

```python

```

**What each part means**

- `.groupby("category")` = split rows into groups.
- `["value"]` = choose the values to summarise.
- `.mean()` = calculate the average for each group.

## Concept 6 - Several Group Summaries At Once With agg()

Sometimes one summary is too thin. `agg()` lets one grouped result show several summary measures at once so the groups are easier to compare.

**Worked dummy example - mean and max together**

```python
dummy_df.groupby("category").agg({"value": ["mean", "max"]})
```

**What each part means**

- `.groupby("category")` = make the groups first.
- `.agg(...)` = ask for several summaries in one result.
- `"value"` = the column being summarised.
- `"mean", "max"` = the summary tools applied to each group.

<div style="border:2px solid #1d4ed8; border-left:8px solid #1d4ed8; padding:12px 14px; border-radius:8px; margin:10px 0;">
<strong>Worked example code cell is directly below.</strong><br>
This blue-marked step is a demonstration. Read it and inspect the output before you answer the Titanic question.
</div>


````python
dummy_df.groupby("category").agg({"value": ["mean", "max"]})
````

## Concept 7 - Turn A Grouped Result Into A Plain-English Finding

A grouped table is evidence, not the final thought. After you create the grouped result, you should say in plain English what the table shows.

**Worked dummy example - grouped result plus finding sentence**

```python
grouped = dummy_df.groupby("category")["value"].mean()
grouped
print("Category A has the highest average value in this dummy table.")
```

**What each part means**

- The grouping line creates the evidence table.
- The printed sentence says, in plain English, what the grouped result shows.
- A grouped result without a sentence is only half-finished analysis.
